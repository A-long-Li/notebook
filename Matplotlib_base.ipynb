### 折线图

通常横轴为特征，纵轴为取值。

```python
import matplotlib as mlp
import matplotlib.pyplot as plt
import numpy as np
```

$linspace(x,y,z)$函数， 产生$x$和$y$之间等间隔的$n$个数，如果$n = 1$，返回结果为$y$。

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

- 绘制二维坐标系，只有当$show$函数被调用才会绘制出图像


```python
plt.plot(x, y)
plt.show()
```

```python
 cosy = np.cos(x)
# y 在上面赋值就是siny
siny = y.copy()
```

- 绘制图像,后面可以继续增加参数来改变图像

```python
plt.plot(x,siny,color = "black",linestyle = "--")
plt.plot(x,cosy)
plt.show()
```

- 手动指定坐标系范围

```python
plt.plot(x,siny,color = "black",linestyle = "--")
plt.plot(x,cosy)
# 两个函数分别指定x,y轴
plt.xlim(-10,10)
plt.ylim(0,1)
#一个函数指定
# plt.axis([x_low,x_high,y_low,y_high])
plt.show()
```

- 增加坐标轴信息,增加图示。增加标题

```python
#增加图示信息
plt.plot(x,siny,label = "sin(x)")
plt.plot(x,cosy,label = "cos(x)")
plt.legend()
#坐标轴信息
plt.xlabel("x_value")
plt.ylabel("y_value")
#标题
plt.title("Test for matplotlib")
plt.show()
```

### 散点图

横轴纵轴都是特征

```python
plt.scatter(x,siny)
plt.scatter(x,cosy)
plt.show()
```

- 修改表示状态 不透明度

```python
# 二维正态分布
x = np.random.normal(0,1,10000)
y = np.random.normal(0,1,10000)
# 0为完全透明1为完全不透明
plt.scatter(x,y,alpha = 0.5)
plt.show()
```
